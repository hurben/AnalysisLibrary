```r
library(ggplot2)
#T-test for comparing two populations

data <- rnorm(n=100, mean = 10, sd = 2)
data2 <- rnorm(n=100, mean = 15, sd = 5)
```

```r
t.test(data, y = data2)
```

```r
density(data)
```

```r
#plot(density(data))
dat <- data.frame(height = c(data, data2), label = rep(c("a", "b")))
#dat
#Plot.
ggplot(dat, aes(x = height, fill = label)) + geom_density(alpha = 0.5) + xlim(0,40)
ggplot(dat, aes(x = height, fill = label)) + geom_dotplot(alpha = 0.5, binwidth = 0.5)
ggplot(dat, aes(x = height, fill = label)) + geom_histogram(alpha = 0.5)
```
