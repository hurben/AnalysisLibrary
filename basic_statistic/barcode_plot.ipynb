```r
preprocess_data_frame <- function(data_frame){
#Overkill function..
#But just incase, I am making it as dataframe
    data_frame = data_frame[with(data_frame, order(-Probablity)),]

    temp_df <- data.frame(matrix(ncol = 2, nrow = 0), stringsAsFactors = F)
    colnames(temp_df) <- c("rank","tp")
    
    for (i in 1:nrow(data_frame)){

        temp_df[nrow(temp_df)+1,] <- c(i, data_frame[i,4])
  
    }   
    return(temp_df)  
}


gain_tp_position <- function(data_list){

    temp_list <- list()
    
    for (i in 1:length(data_list)){
        if (data_list[i] == 1){
            
            temp_list <- c(temp_list, i)
            
        }
    }
    temp_list <- as.numeric(temp_list)
    return(temp_list)
}
```
